# Tesla stocks :

```python
# importing necessary libraries
import pandas as pd
import yfinance as yf
```

```python
# Define the ticker symbol for Tesla
ticker = 'TSLA'

# Use yfinance to download the stock data (example: from 2020 to 2024)
tesla_data = yf.download(ticker, start="2020-01-01", end="2024-01-01")

# Reset the index to move the 'Date' column from the index to a regular column
tesla_data_reset = tesla_data.reset_index()

# Save the DataFrame to a CSV file
tesla_data_reset.to_csv('tesla_stock_data.csv', index=True)

# Display the first five rows of the reset DataFrame
print(tesla_data_reset.head())
```

```python
import matplotlib.pyplot as plt

# Plot Tesla's closing price over time
tesla_data['Close'].plot(figsize=(10, 6))
plt.title('Tesla Stock Closing Price')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.show()
```

# GME Revenue :

```python
# Use Webscraping to Extract GME Revenue Data Display the last five rows of the gme_revenue dataframe using the tail function.

import requests
from bs4 import BeautifulSoup
import pandas as pd

# URL of the page to scrape (replace with the actual URL)
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/stock.html"

# Send an HTTP request to the website
response = requests.get(url)

# Parse the page content
soup = BeautifulSoup(response.content, 'html.parser')

# Extract the relevant table (adjust the selection according to the webpage structure)
table = soup.find('table')  # Find the first table or adjust if multiple tables exist

# Convert the table to a pandas dataframe
gme_revenue = pd.read_html(str(table))[0]

# Display the last five rows
print(gme_revenue.tail())

# Optionally, save the dataframe to a CSV file
gme_revenue.to_csv("gme_revenue.csv", index=False)
```

```python
# Use the make_graph function to graph the Tesla Stock Data, also provide a title for the graph.

import yfinance as yf
import matplotlib.pyplot as plt

# Function to make the graph
def make_graph(stock_data, title):
    plt.figure(figsize=(10,6))
    plt.plot(stock_data['Date'], stock_data['Close'], label='Close Price')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Stock Price (USD)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Download Tesla stock data
tesla_data = yf.download("TSLA", start="2010-01-01", end="2024-01-01")

# Reset the index so 'Date' is a column
tesla_data.reset_index(inplace=True)

# Call the make_graph function
make_graph(tesla_data, "Tesla Stock Price Over Time")
```

```python
# Use the make_graph function to graph the GameStop Stock Data, also provide a title for the graph.

import yfinance as yf
import matplotlib.pyplot as plt

# Function to make the graph
def make_graph(stock_data, title):
    plt.figure(figsize=(10,6))
    plt.plot(stock_data['Date'], stock_data['Close'], label='Close Price', color='blue')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Stock Price (USD)')
    plt.legend()
    plt.grid(True)
    plt.show()

# Download GameStop stock data
gme_data = yf.download("GME", start="2010-01-01", end="2024-01-01")

# Reset the index so 'Date' is a column
gme_data.reset_index(inplace=True)

# Call the make_graph function for GameStop stock data
make_graph(gme_data, "GameStop Stock Price Over Time")
```
