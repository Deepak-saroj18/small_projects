## Principal Component Analysis (PCA):ace.

#### Perform PCA to reduce the dimensionality of the dataset. Plot the explained variance ratio for each principal component.
#### Visualize the data in the new principal component space.

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import pandas as pd

# Load dataset
iris = sns.load_dataset('iris')

# ✅ Select only numeric features for PCA
X = iris.drop('species', axis=1)   # keep only SepalLength, SepalWidth, PetalLength, PetalWidth
y = iris['species']                # target labels for coloring

# ✅ Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ✅ Perform PCA
pca = PCA()
X_pca = pca.fit_transform(X_scaled)

# ✅ Explained variance ratio
explained_var = pca.explained_variance_ratio_

plt.figure(figsize=(8,5))
plt.plot(range(1, len(explained_var)+1), explained_var, marker='o', linestyle='--')
plt.title("Explained Variance Ratio by Principal Components")
plt.xlabel("Principal Component")
plt.ylabel("Explained Variance Ratio")
plt.show()

# ✅ 2D Visualization in PCA space
plt.figure(figsize=(8,6))
plt.scatter(X_pca[:,0], X_pca[:,1], c=y.astype('category').cat.codes, cmap='viridis')
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("Iris Data Projected onto First Two Principal Components")
plt.colorbar(label="Species (encoded)")
plt.show()
```

## Feature Importance:

#### Use a machine learning model (e.g., Random Forest) to determine the importance of each feature. Plot the feature importances to identify the most significant features.

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

```python
iris.columns
```

```python
x = iris[['PetalLengthCm', 'SepalLengthCm', 'SepalWidthCm']]
y = iris[['PetalWidthCm']]
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=2529)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(random_state=2529)
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score

# Evaluate regression
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("✅ Mean Squared Error:", mse)
print("✅ R² Score:", r2)
```

```python
# Z-score and IQR
```

```python
# ✅ Keep only numeric columns
iris_num = iris.drop('species', axis=1)

# ------------------------
# 📌 Z-SCORE METHOD
# ------------------------
from scipy.stats import zscore

z_scores = np.abs(zscore(iris_num))
threshold = 3  # common cutoff
outliers_z = (z_scores > threshold).any(axis=1)

print("Number of outliers detected using Z-score:", outliers_z.sum())

# ------------------------
# 📌 IQR METHOD
# ------------------------
Q1 = iris_num.quantile(0.25)
Q3 = iris_num.quantile(0.75)
IQR = Q3 - Q1

outliers_iqr = ((iris_num < (Q1 - 1.5 * IQR)) | (iris_num > (Q3 + 1.5 * IQR))).any(axis=1)

print("Number of outliers detected using IQR:", outliers_iqr.sum())

# ------------------------
# 📊 Visualize outliers
# ------------------------
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.boxplot(data=iris_num, orient="h")
plt.title("Boxplots (IQR-based Outlier Visualization)")
plt.show()
```
